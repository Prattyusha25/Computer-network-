```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck's constant in Js
m_e = 9.109e-31  # Mass of an electron in kg
eV_conversion = 1.602e-19  # Energy conversion factor from Joules to eV

# Function to calculate energy levels for a given length and quantum number n
def energy_levels(n, L_nm):
    L_m = L_nm * 1e-9  # Convert nm to meters
    energy_joules = (n**2 * h**2) / (8 * m_e * L_m**2)  # Energy in Joules
    energy_ev = energy_joules / eV_conversion  # Convert to eV
    return energy_ev

# Quantum numbers to calculate (ground state and first 2 excited states)
quantum_numbers = [1, 2, 3]

# Dimensions of the quantum well/quantum dot/quantum wire in nm
lengths_nm = [1, 3, 5]

# Calculate and plot energy levels
def plot_energy_levels(structure_type):
    plt.figure(figsize=(8, 6))

    for L_nm in lengths_nm:
        energies = []
        for n in quantum_numbers:
            energies.append(energy_levels(n, L_nm))
        plt.plot(quantum_numbers, energies, marker='o', label=f'L = {L_nm} nm')

    plt.title(f'Energy Levels for {structure_type} (in eV)')
    plt.xlabel('Quantum Number (n)')
    plt.ylabel('Energy (eV)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot energy levels for quantum well, dot, and wire
plot_energy_levels("Quantum Well")
plot_energy_levels("Quantum Dot")
plot_energy_levels("Quantum Wire")
```
